```python
import matplotlib.pyplot as plt
```

```python
df_survive = df.groupby('Survived').size().reset_index(name = 'Numbers')
df_survive
```

```python
x = df_survive['Survived']
y = df_survive['Numbers']
plt.pie(y, labels = x,autopct='%.2f%%',startangle=90)
plt.show()
```

```python
plt.figure(figsize = (3, 5))
plt.xlabel('Situation')
plt.ylabel('Numbers of people')
x = df_survive['Survived']
y = df_survive['Numbers']
plt.bar(x, y)
plt.grid(linestyle = ':', which = 'both')
plt.show()
```
