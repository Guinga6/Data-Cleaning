```python
# changing the value name 
data['platforms'].replace('mac;linux', 'macbook and linux', inplace=True)
data['platforms'].replace('windows;linux', 'windows and linux', inplace=True)
data['platforms'].replace('windows;mac;linux', 'windows and macbook and linux', inplace=True)
data['platforms'].replace('windows;mac', 'windows and macbook', inplace=True)
```

```python
# a look in the platforms column
data_platforms = data.groupby('platforms').size().reset_index( name = 'Numbers')
data_platforms.head()
```

```python
# Let's make this data into a graph.
x = data_platforms['platforms']
y = data_platforms['Numbers']
plt.pie(y, labels = x, autopct='%.0f%%',startangle = 100 )
plt.grid()
plt.show()
```
